```python
from ulbc import *
```

```python
us = [x,y]
k=RIF(0.8); b=RIF(0.6); c=RIF(0.3); e=RIF(0.05); f=RIF(2)
vf = vector([b*x*(RIF(1)-x) - c*x*(k-x)*y,
            -e*y*(RIF(1)+y)+f*x*(k-x)*y])
v0 = [RIF(0.5), RIF(2.0)]
v1 = [RIF(0.4), RIF(2.5)]
v2 = [RIF(1,1.2), RIF(4,6)]
```

```python
p = streamline_plot(tuple(vf),(x,0,1.5),(y,0,8))\
    + point(v0, size=100, marker='X', color='red')\
    + point(v1, size=100, marker='X', color='green')\
    + polygon([(v2[0].lower(), v2[1].lower()),
               (v2[0].lower(), v2[1].upper()),
               (v2[0].upper(), v2[1].upper()),
               (v2[0].upper(), v2[1].lower())],
              color='orange')
p.set_aspect_ratio(1/6)
p.show()
```

```python
P = Atomic((x - 1)^2 + y^2 - 0.5)
```

```python
sig = P.signal_for_system(vf, v2, 5, order=5, step=(0.001, 0.05),
                    precondition=1,
                    estimation=1e-3,
                    integrationScheme=2,
                    cutoff_threshold=1e-5)
sig.plot()
```

```python
reach = Reach(vf, v2, 10, order=5, step=0.01,
              precondition=1,
              estimation=1e-3,
              integrationScheme=2,
              cutoff_threshold=1e-7)
```

```python
reach.wplot('x', 'y')
```

```python
reach.sage_interval_plot('x', 'y',
                         poly=(x - 1)^2 + y^2 - 0.5,
                         aspect_ratio='automatic')
```

```python
roots = reach.roots(P.p, P.dpdt(vf))
```

```python
[root.str(style='brackets') for root in roots]
```
