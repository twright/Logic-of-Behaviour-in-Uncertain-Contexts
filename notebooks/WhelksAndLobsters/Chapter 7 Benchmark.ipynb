```python
from whelks_and_lobsters_benchmarking import *
from lbuc import RestrictionMethod
```

```python
n = 7
```

# Property

```python
P = Atomic((var("Whelk") - 1)**2 + var("Lobster")**2 < 0.2)
```

```python
Qbox = [RIF(0, 0.31), RIF(1.5, 4)]

Q = (  Atomic(var("Whelk") < Qbox[0].upper())
     & Atomic(var("Whelk") > Qbox[0].lower())
     & Atomic(var("Lobster") < Qbox[1].upper())
     & Atomic(var("Lobster") > Qbox[1].lower()) )

prop1 = F(RIF(0, 10), Q)
```

```python
prop2 = F(RIF(0, 5), G(RIF(0, 5), P))
```

# Experiments

# Results

```python
ress
```

```python
def results(key):
    result_list = []
 
    for res in ress[key]:
        try:
            result_list.append(
                (res['depth'], *res['future'].result(timeout=0))
            )
        except:
            pass

    return result_list
```

```python
resu = results('p1prop1'); resu
```

```python
pd.DataFrame({
    'depth': [res[0] for res in resu],
    'total': [res[1] for res in resu],
    'num_flowpipes': [res[2].total_count('Running Flow*')
                      for res in resu],
    'composition': [res[2].total_time('composing flowpipe')
                    for res in resu],
    'flowstar': [res[2].total_time('Running Flow*')
                 for res in resu],
    'monitoring': [
        res[1]
            - res[2].total_time('Running Flow*')
            - res[2].total_time('composing flowpipe')
        for res in resu
    ]
})
```

```python
ress.keys()
```

```python
columns = ('params', 'prop', 'variant', 'depth', 'total', 'num_flowpipes', 'flowstar',
           'composition', 'monitoring')

df = pd.DataFrame(
    sum(
        ([
            (
                key[0:2],
                key[2:7],
                key[7:],
                res[0],
                res[1],
                res[2].total_count('Running Flow*'),
                res[2].total_time('Running Flow*'),
                res[2].total_time('composing flowpipe'),
                res[1]
                    - res[2].total_time('Running Flow*')
                    - res[2].total_time('composing flowpipe'),
            ) for res in results(key)]
         for key in ress.keys() if len(results(key)) > 0),
        [],
    ),
    columns=columns,
)
df
```
