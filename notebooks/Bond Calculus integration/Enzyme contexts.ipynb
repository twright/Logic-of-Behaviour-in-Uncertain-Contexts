## Delayed Enzyme Introduction via Context

```python
r1 = sys1.reach(5)
```

```python
sys3 = to_context_body("[0.125 .. 0.175] E").apply_jump(sys1.with_y0(r1(RIF(5, 5))))
sys3
```

```python
r2 = sys3.reach(10)
```

```python
r2.successful
```

```python
graphics_array([
  r1.sage_plot(var("S"), color=["blue", "blue"], fillcolor="blue", fill={0: [1]}, ymax=1),
#+ r1.sage_plot(var("P"), color=["orange", "orange"], fillcolor="orange", fill={0: [1]}),
  r2.sage_plot(var("S"), color=["blue", "blue"], fillcolor="blue", fill={0: [1]}, ymax=1)\
+ r2.sage_plot(var("P"), color=["orange", "orange"], fillcolor="orange", fill={0: [1]})
], 1, 2)
```

```python
prop4 = "[0.125 .. 0.175] E" >> F(RIF(0, 5), Atomic(var("P") > 0.125)); prop4
```

```python
sig4 = prop4.signal_for_system(sys1, 10)
```

```python
sig4.plot()
```

## Introducing Enzyme Affinity Rule

```python
pi3 = model.process("[1.0] S || [0.1 .. 0.15] E with network M4(0.0, 0.1, 0.5, 0.1)"); pi3
```

```python
sys4 = pi3.as_system; sys4
```

```python
r = sys4.reach(20)
```

```python
r.sage_plot(var("S"), color=["blue", "blue"], fillcolor="blue", fill={0: [1]})
```

```python
r1 = sys4.reach(5)
```

```python
sys5 = to_context_body("{ e || s at rate MA([1.0 .. 1.1]); }").apply_jump(sys4.with_y0(r1(RIF(5, 5))))
sys5
```

```python
r2 = sys5.reach(10)
```

```python
r2.successful
```

```python
graphics_array([
  r1.sage_plot(var("S"), color=["blue", "blue"], fillcolor="blue", fill={0: [1]}, ymax=1)\
+ r1.sage_plot(var("E"), color=["green", "green"], fillcolor="green", fill={0: [1]}),
  r2.sage_plot(var("S"), color=["blue", "blue"], fillcolor="blue", fill={0: [1]}, ymax=1)\
+ r2.sage_plot(var("P"), color=["orange", "orange"], fillcolor="orange", fill={0: [1]})\
+ r2.sage_plot(var("E"), color=["green", "green"], fillcolor="green", fill={0: [1]}),
], 1, 2)
```

```python
prop5 = "{ e || s at rate MA(1.0); }" >> F(RIF(0, 5), Atomic(var("P") > 0.1)); prop5
```

```python
sig5 = prop5.signal_for_system(sys4, 10)
```

```python
sig5.plot()
```

```python
sig7 = prop5.signal_for_system(sys4, 10, use_masks=True, symbolic_composition=True)
```
