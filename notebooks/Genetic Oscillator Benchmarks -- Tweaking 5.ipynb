```python
def extract_result(d):
    return {
        'prop': d['prop'],
        'duration': d['duration'],
        'kwargs': d['kwargs'],
        'result': d['future'].result(),
    }

def extract_results(ds):
    return list(map(extract_result, (d[0] for d in ds)))
```
