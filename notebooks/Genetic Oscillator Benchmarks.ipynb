# Analysis

```python
import pandas as pd
```

```python
def total_time_for_key(inst, ks):
    if isinstance(ks, str):
        ks = [ks]

    return sum(inst.total_time(k)
               for k in ks)


def total_times(results, keys, keys_short=None):
    ress = [
        f.result() for f in futures(results)
    ]
    if keys_short is None:
        keys_short = keys
    data = pd.DataFrame({
        short_key: [total_time_for_key(r[2], key) for r in ress]
        for short_key, key in zip(keys_short, keys)
    })
    data.insert(len(keys), 'Total',
                [r[1] for r in ress])
    return data
```

## Export Data

```python
series = [
    'masked',
    'masked_no_symbolic',
    'masked_no_symbolic_no_selective',
    'masked_no_selective',
    'masked_symbolic_low_order',
    'unmasked',
    'unmasked_no_symbolic',
    'upfront',
    'upfront_selective',
    'masked_reversed',
    'unmasked_reversed',
]
# total_timess = {
#    s: pd.load_csv(f'data/analysis3_raw_{s}.csv')
#    for s in series 
# }
total_timess = {
    s: total_times(eval(f'results_{s}'), cols_full, cols_short)
    for s in series 
}
```

```python
import pickle
```

```python
for s in series:
    with open(f'data/results_{s}.pkl', 'wb') as file:
        pickle.dump([f.result() for f in futures(eval(f'results_{s}'))],
                    file)
```

```python
for s, d in total_timess.items():
    d.to_csv(f'data/benchmark_{s}.csv', index_label=False)
```
