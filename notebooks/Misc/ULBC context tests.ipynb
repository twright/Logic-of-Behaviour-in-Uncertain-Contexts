```python
from flowstar import *
from lbuc import *
from flowstar.poly import Poly
```

```python
R.<x,y> = RIF[]
```

```python
odes = [
    -y,
    x,
]
initials = [(4,5), (1,2)]
reach = Reach(odes, [(4,5), (1,2)], 2*pi, 0.1, order=10)
```

```python
def plot_signal(sig, tick=pi/3):
    return sig.to_domain(RIF(0, sig.domain.upper())).plot().show(
        ticks=[tick, 1], tick_formatter=pi,
        aspect_ratio=(1/8)*sig.domain.upper())
```

```python
reach.wplot('x', 'y')
```

```python
reach.sage_interval_plot('x', 'y')
```

```python
reach.sage_interval_plot('x', 'y', poly=y)
```

```python
reach2 = Reach(odes,
               [z + RIF(4, 5) if i == 1 else z
                for (i, z)
                in enumerate(reach(RIF(pi/3, pi/2)), 1)],
               pi, (0.001, 1), order=5)
```

```python
reach2.sage_interval_plot('x', 'y', poly=y)
```

```python
P = {x: RIF(4, 5)} >> G(RIF(0, pi/6), Atomic(y))
P
```

```python
reach3 = Reach(odes,
               P.context_jump(reach(RIF(pi/3,pi/2))),
               pi/6 + 1e-3)
```
