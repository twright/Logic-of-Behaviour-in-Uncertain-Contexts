## Example 1: Simple

```python
from flowstar.reachability import Reach
from flowstar.poly import Poly
from lbuc import *
```

```python
R.<x,y> = RIF[]
```

```python
odes = [
    -y,
    x,
]
initials = [RIF(4,5), RIF(1,2)]
reach = Reach(odes, initials, 2*pi, 0.1, order=10)
```

```python
def plot_signal(sig):
    return sig.to_domain(RIF(0,2*pi)).plot().show(
        ticks=[pi/2, 1], tick_formatter=pi, aspect_ratio=2/3)
```

```python
reach.res
```

```python
P = Atomic(x - 3); P
```

```python
P.context_signal(reach, odes, initials).signal.plot() + P.sage_plot(reach)
```

```python
plot_signal( P.context_signal_for_system(odes, initials, 2*pi).signal )
```

```python
sig = P.context_signal_for_system(odes, initials, 2*pi).refined_signal(2)
```

```python
plot_signal( sig )
```

```python
Q = Atomic(3 - y); Q
```

```python
Q.signal(reach, odes).plot() + Q.sage_plot(reach)
```

```python
plot_signal( Q.context_signal_for_system(odes, initials, 2*pi).signal )
```

```python
plot_signal( (P & Q).context_signal_for_system(odes, initials, 2*pi).signal )
```

```python
sig = (P & Q).context_signal_for_system(odes, initials, 2*pi).refined_signal(1)
```

```python
plot_signal(sig)
```

```python
sig1 = (P | Q).context_signal_for_system(odes, initials, 2*pi).refined_signal(1)
```

```python
sig2 = ( P.context_signal_for_system(odes, initials, 2*pi).refined_signal(1)
       | Q.context_signal_for_system(odes, initials, 2*pi).refined_signal(1))
```

```python
plot_signal(sig1)
```

```python
plot_signal(sig2)
```

```python
plot_signal( G(RIF(1,1.2), P).signal_for_system(odes, initials, 2*pi) )
```

```python
plot_signal( F(RIF(3.5, 4), Q).signal_for_system(odes, initials, 2*pi) )
```

```python
plot_signal( (G(RIF(1,1.2), P) | F(RIF(3.5,4), Q)).signal_for_system(odes, initials, 2*pi) )
```

```python
(G(RIF(1,1.2), P) & F(RIF(3.5,4), Q)).signal_for_system(odes, initials, 2*pi).plot()
```
