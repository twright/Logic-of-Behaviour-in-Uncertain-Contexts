# Profile signal monitoring

```python
import pstats
import cProfile
```

```python
R.<x,y> = QQ[]
from lbuc import *
us = [x,y]
k=RIF(0.8); b=RIF(0.6); c=RIF(0.3); e=RIF(0.05); f=RIF(2)
vf = vector([b*x*(RIF(1)-x) - c*x*(k-x)*y,
            -e*y*(RIF(1)+y)+f*x*(k-x)*y])
v0 = [RIF(0.5), RIF(2.0)]
v1 = [RIF(0.4), RIF(2.5)]
v2 = [RIF(1,1.2), RIF(4,6)]
v3 = v2
```

```python
P = Atomic((x - 1)^2 + y^2 - 0.2)
prop1 = F(RIF(4, 4.5), P)
```

```python
def work1(prop, n=5):
    return prop.signal_for_system(vf, v3, n, order=5, step=0.01,
                    precondition=1,
                    estimation=1e-3,
                    integrationScheme=2,
                    cutoff_threshold=1e-5,
                    epsilon_ctx=0.1,
                    precompose_taylor_models=True)
```

```python
def work2(prop, n=5):
    return prop.signal_for_system(vf, v3, n, order=5, step=0.01,
                    precondition=1,
                    estimation=1e-3,
                    integrationScheme=2,
                    epsilon_ctx=0.1,
                    cutoff_threshold=1e-5,
                    use_masks=True)
```

```python
def work3(prop, n=5):
    return prop.signal_for_system(vf, v3, n, order=5, step=0.01,
                    precondition=1,
                    estimation=1e-3,
                    symbolic_composition=True,
                    tentative_unpreconditioning=False,
                    epsilon_ctx=0.1,
                    integrationScheme=2,
                    cutoff_threshold=1e-5)
```

```python
def work4(prop, n=5):
    return prop.signal_for_system(vf, v3, n, order=5, step=0.01,
                    precondition=1,
                    estimation=1e-3,
                    integrationScheme=2,
                    cutoff_threshold=1e-5,
                    epsilon_ctx=0.1,
                    tentative_unpreconditioning=False,
                    symbolic_composition=True,
                    use_masks=True)
```

```python
def work5(prop, n=5):
    return prop.signal_for_system(vf, v3, n, order=5, step=0.01,
                    precondition=1,
                    estimation=1e-3,
                    integrationScheme=2,
                    cutoff_threshold=1e-5,
                    epsilon_ctx=0.1,
                    tentative_unpreconditioning=True,
                    symbolic_composition=True,
                    use_masks=True)
```

```python
for i in range(1, 6):
    profile_name = 'profile work{}.prof'.format(i)
    cProfile.runctx('work{}(prop1)'.format(i),
                    globals(),
                    locals(),
                    profile_name)
    s = pstats.Stats(profile_name)
    print("=== Profile work{} ===".format(i))
    s.strip_dirs().sort_stats("time").print_stats()
```
