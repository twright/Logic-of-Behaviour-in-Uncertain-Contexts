# Lotka Volterra predator prey

```python
import sympy as sym
import numpy as np
from sympy.abc import t
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib notebook
sym.init_printing()
```

```python
xs = [sym.Function('"Rabbit"'),sym.Function('"Fox"')]
xts = [x(t) for x in xs]
odes = [sym.Eq(xts[0].diff(), sym.simplify((xts[0]) * (2.0) + (-1.0) * ((xts[0]) * ((xts[1]) * (0.5))))), sym.Eq(xts[1].diff(), sym.simplify((xts[0]) * ((xts[1]) * (0.5)) + (-1.0) * ((xts[1]) * (0.8))))]
y0 = [2.0, 1.0]
ts = np.linspace(0.0,20.0,1000)
rhss = [eqn.rhs for eqn in odes]
Jac = sym.Matrix(rhss).jacobian(xts)
f = sym.lambdify(xts, rhss, modules='numpy')
J = sym.lambdify(xts, Jac, modules='numpy')
```

```python
ys = odeint(lambda xs,_: f(*xs), y0, ts, (), J)
```

```python
plt.figure()
plt.plot(ts,ys)
```

```python
Y,X = np.mgrid[0:15:100j,0:10:100j]
U,V = f(X,Y)
mag = np.sqrt(U**2 + V**2)
```

```python
plt.figure()
plt.streamplot(X,Y,U,V,color=mag,density=1.5)
plt.colorbar()
```

# Lotka-Volterra interguild

```python
import sympy as sym
import numpy as np
from sympy.abc import t
from scipy.integrate import odeint

xs = [sym.Function('"Blue"'),sym.Function('"Red"')]
xts = [x(t) for x in xs]
odes = [sym.Eq(xts[0].diff(), sym.simplify((10.0) * ((xts[0]) * (0.1)) + ((xts[0]) * (-1.0)) * ((xts[0]) * (0.1)) + ((xts[0]) * (0.1)) * ((-1.0) * ((xts[1]) * (0.5))))), sym.Eq(xts[1].diff(), sym.simplify((15.0) * ((xts[1]) * (6.666666666666667e-2)) + ((xts[1]) * (-1.0)) * ((xts[1]) * (6.666666666666667e-2)) + ((xts[1]) * (6.666666666666667e-2)) * ((-1.0) * ((xts[0]) * (0.5)))))]
rhss = [eqn.rhs for eqn in odes]
Jac = sym.Matrix(rhss).jacobian(xts)
f = sym.lambdify(xts, rhss, modules='numpy')
J = sym.lambdify(xts, Jac, modules='numpy')
```

```python
y0 = [1.0, 1.0]
ts = np.linspace(0.0,20.0,1000)
ys = odeint(lambda xs,_: f(*xs), y0, ts, (), J)
plt.figure()
plt.plot(ts,ys)
```

```python
import sympy as sym
import numpy as np
from sympy.abc import t
from scipy.integrate import odeint

xs = [sym.Symbol('Blue'),sym.Symbol('Red')]
xts = [x for x in xs]
odes = [sym.Eq(xts[0].diff(), sym.simplify((15.0) * ((xts[0]) * (0.14666666666666667)) + ((xts[0]) * (-1.0)) * ((xts[0]) * (0.14666666666666667)) + ((xts[0]) * (0.14666666666666667)) * ((-1.0) * ((xts[1]) * (0.1))) + (-1.0) * ((xts[0]) * (xts[1])))), sym.Eq(xts[1].diff(), sym.simplify((xts[1]) * (2.0) + (xts[1]) * ((xts[1]) * (-1.0)) + (xts[1]) * ((-1.0) * ((xts[0]) * (0.6))) + (xts[0]) * ((xts[1]) * (0.5))))]
y0 = [1.0, 1.0]
ts = np.linspace(0.0,50.0,1000)
rhss = [eqn.rhs for eqn in odes]
Jac = sym.Matrix(rhss).jacobian(xts)
f = sym.lambdify(xts, rhss, modules='numpy')
J = sym.lambdify(xts, Jac, modules='numpy')
ys = odeint(lambda xs,_: f(*xs), y0, ts, (), J)
```

## Differential geometry

```python
import sympy.diffgeom as dg
```

```python
ddt = rhss[0]*eR + rhss[1]*eB
```

```python
R, B = rect.coord_functions()
```

```python
rhsss = f(R,B)
ddt = rhsss[0]*eR + rhsss[1]*eB
```

```python
ddt
```

```python
factsb = [fact.subs({xts[0]:R,xts[1]:B}) for fact in facts]
```

```python
dg.LieDerivative(ddt,factsb[1])
```

## Plot lie derivatives

```python
p = Plot(xlim = (0,xmax), ylim=(0,ymax), xlabel="R", ylabel="B")
p._backend = p.backend(p)
fig = p._backend.fig
for fact in facts:
    p.extend(sym.plot_implicit(fact,x_var=(xts[0],0,xmax), y_var=(xts[1],0,ymax),show=False,line_color="red"))
for fact in factsb:
    lderiv = dg.LieDerivative(ddt,fact).subs({R:xts[0],B:xts[1]})
    p.extend(sym.plot_implicit(lderiv,x_var=(xts[0],0,xmax), y_var=(xts[1],0,ymax),show=False,line_color="red"))
plt.streamplot(X,Y,U,V,color=mag,density=2,linewidth=np.log(1.3+6*mag/mag.max()))
p._backend.show()
```

# Predator prey role reversal

```python
import sympy as sym
import numpy as np
from sympy.abc import t
from scipy.integrate import odeint

xs = [sym.Symbol('R',real=True),sym.Symbol('B',real=True)]
xts = [x for x in xs]
odes = [sym.Eq(xts[0].diff(), sym.simplify((xts[0]) * (0.6) + ((xts[0]) * (-1.0)) * ((xts[0]) * (0.6)) + (-1.0) * ((xts[1]) * ((0.8) * ((xts[0]) * (0.3)))) + (-1.0) * ((xts[1]) * (((xts[0]) * (-1.0)) * ((xts[0]) * (0.3)))))), sym.Eq(xts[1].diff(), sym.simplify((xts[1]) * ((0.8) * ((xts[0]) * (2.0))) + (xts[1]) * (((xts[0]) * (-1.0)) * ((xts[0]) * (2.0))) + (-1.0) * ((xts[1]) * (5.0e-2)) + (-1.0) * ((xts[1]) * ((xts[1]) * (5.0e-2)))))]
y0 = [0.7, 7.0]
ts = np.linspace(0.0,50.0,1000)
rhss = [eqn.rhs for eqn in odes]
Jac = sym.Matrix(rhss).jacobian(xts)
f = sym.lambdify(xts, rhss, modules='numpy')
J = sym.lambdify(xts, Jac, modules='numpy')
```

## Sample simulation

```python
y0 = [0.1, 1]
ts = np.linspace(0.0,50,1000)
ys = odeint(lambda xs,_: f(*xs), y0, ts, ())
plt.figure()
plt.plot(ts,ys)
```

```python
def lie(f,n):
    for i in range(n):
        f = dg.LieDerivative(ddt,f)
    return f
```
