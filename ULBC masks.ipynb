## Example 1: Simple

```python
from flowstar.reachability import Reach
from flowstar.poly import Poly
from ulbc import *
```

```python
R.<x,y> = RIF[]
```

```python
odes = [
    -y,
    x,
]
initials = [(4,5), (1,2)]
reach = Reach(odes, [(4,5), (1,2)], 2*pi, 0.1, order=10)
```

```python
def plot_signal(sig):
    return sig.to_domain(RIF(0,2*pi)).plot().show(
        ticks=[pi/2, 1], tick_formatter=pi, aspect_ratio=2/3)
```

```python
reach.res
```

```python
reach(RIF(1,2))[0].str(style='brackets')
```

```python
reach.sage_plot(x, (0.9,2.1))
```

```python
reach(RIF(1,2))[1].str(style='brackets')
```

```python
reach.sage_plot(y, (1,2))
```

```python
P = Atomic(x - 3); P
```

```python
P.signal(reach, odes).plot() + P.sage_plot(reach)
```

```python
plot_signal( P.signal_for_system(odes, initials, 2*pi) )
```

```python
Q = Atomic(3 - y); Q
```

```python
Q.signal(reach, odes).plot() + Q.sage_plot(reach)
```

```python
plot_signal( Q.signal_for_system(odes, initials, 2*pi) )
```

### PolyObserver

```python
from flowstar.observers import PolyObserver
```

```python
observer = PolyObserver(x - 3, reach, False); observer
```

```python
roots = observer.roots()
print '[', '\n, '.join(root.str(style='brackets') for root in roots), ']'
```

```python
observer_masked = PolyObserver(x - 3, reach, False, mask); observer_masked
```

```python
roots = observer_masked.roots()
print '[', '\n, '.join(root.str(style='brackets') for root in roots), ']'
```

```python
observer.check(RIF(3*pi/2))
```

```python
observer_masked.check(RIF(3*pi/2))
```
