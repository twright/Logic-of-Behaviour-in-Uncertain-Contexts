```python
from flowstar.reachability import Reach, Poly
```

## Example 1

```python
varsc = [b't', b'x']
pt = Poly(1, 't', 0, varsc)
px = Poly(-1, 'x', 1, varsc)
print "t' = {}\nx' = {}".format(pt, px)
```

```python
C = Reach(
    varsc,
    [pt, px],
    [(0,0), (5,5)],
    5,
    0.5,
    order=10,
)
```

```python
C.res
```

```python
C.num_flowpipes
```

```python
C.wplot('t', 'x')
```

```python
C.sage_plot_manual('x', 0.05)
```

```python
C.sage_plot('x', 0.05)
```

```python
_, (lo, hi) = C(RIF(0.2,1.1))
```

```python
   C.sage_plot('x', 0.05) \
 + C.sage_plot('x') \
 + line([(0.2, lo), (1.1, lo)]) \
 + line([(0.2, hi), (1.1, hi)], color='#3bcc00')
```

## Example 2

```python
varsd = [b't', b'x', b'y']
qt = Poly(1, 't', 0, varsd)
qx = Poly(-1, 'y', 1, varsd)
qy = Poly(1, 'x', 1, varsd)
print "t' = {}\nx' = {}\ny' = {}".format(qt, qx, qy)
```

```python
D = Reach(
    varsd,
    [qt, qx, qy],
    [(0,0), (4,5), (1,2)],
    10,
    0.1,
    order=5,
)
```

```python
D.res
```

```python
D.wplot(b't', b'x')
```

```python
D.sage_plot('x')
```

```python
D.wplot(b't', b'y')
```

```python
D.sage_plot('y')
```

```python
D.wplot(b'x', b'y')
```
