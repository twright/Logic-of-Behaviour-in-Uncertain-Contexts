```python
from flowstar.reachability import Reach, Poly
```

## Example 1

```python
varsc = [b't', b'x']
pt = Poly(1, 't', 0, varsc)
px = Poly(-1, 'x', 1, varsc)
print "t' = {}\nx' = {}".format(pt, px)
```

```python
C = Reach(
    varsc,
    [pt, px],
    [(0,0), (5,5)],
    5,
    0.1,
    order=3,
)
```

```python
C.res
```

```python
C.num_flowpipes
```

```python
C.wplot('t', 'x')
```

## Example 2

```python
varsd = [b't', b'x', b'y']
qt = Poly(1, 't', 0, varsd)
qx = Poly(-1, 'y', 1, varsd)
qy = Poly(1, 'x', 1, varsd)
print "t' = {}\nx' = {}\ny' = {}".format(qt, qx, qy)
```

```python
D = Reach(
    varsd,
    [qt, qx, qy],
    [(0,0), (4,5), (1,2)],
    10,
    0.1,
    order=5,
)
```

```python
D.res
```

```python
D.wplot(b't', b'x')
```

```python
D.wplot(b't', b'y')
```

```python
D.wplot(b'x', b'y')
```

```python
D((0.13, 0.95))
```

```python
p = Graphics()
step = 0.1
lo1 = hi1 = 0
for i in range(0,100):
    t = step*i
    (_, (lo, hi), _) = D((t,t+step))
    p += line([(t, lo), (t+step, lo)])
    p += line([(t, lo1), (t, lo)])
    p += line([(t, hi), (t+step, hi)], color='green')
    p += line([(t, hi1), (t, hi)], color='green')
    lo1 = lo
    hi1 = hi
    
p
```

```python
D.wplot(b't', b'x')
```
