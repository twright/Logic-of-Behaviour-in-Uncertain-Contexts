```python
from flowstar.reachability import Reach, Poly
```

```python
varsc = [b't', b'x']
pt = Poly(1, 't', 0, varsc)
px = Poly(-1, 'x', 1, varsc)
print "t' = {}\nx' = {}".format(pt, px)
```

```python
C = Reach(
    varsc,
    [pt, px],
    [(0,0), (5,5)],
    5,
    0.1,
    order=3,
)
```

```python
C.res
```

```python
C.num_flowpipes
```

```python
C.wplot('t', 'x')
```

```python
varsd = [b't', b'x', b'y']
qt = Poly(1, 't', 0, varsd)
qx = Poly(-1, 'y', 1, varsd)
qy = Poly(1, 'x', 1, varsd)
print "t' = {}\nx' = {}\ny' = {}".format(qt, qx, qy)
```

```python
D = Reach(
    varsd,
    [qt, qx, qy],
    [(0,0), (4,5), (1,2)],
    10,
    0.1,
    order=5,
)
```

```python
D.res
```

```python
D.wplot(b'x', b'y')
```
